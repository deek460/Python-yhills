## statistics

### inferntial -> portion(sample)

## measures of central tendency -> mean median mode *outlier*
### bcz of outliers mean can change a lot so it is not wise to replace missing data with mean
### median is better for replacing missing data
### catagorical data cant have mean or median but it can have mode
### sometimes we can have more than one mode and that becomes a challenge
## measures of dispersion ->  range, variance, std deviation
### rnge maxval-minval
### variance and std deviation -> measure of spread of data
### (sigma)^2 sum of x-mean ka mod ka square divided by n
### std deviation is sigma
## covariance and correlation
### relation b/w diff attributes
### covariance = sum of [(x-mean of x)(y-mean of y)]/n
### if both x and y change in same direction covariance is positive
### if x and y change in opposite direction the covariance is negetive
### corelation = [covar(x,y)]/(sigma-x . sigma-y)
### correlation lies in [-1,1]
### if correlation is +1 relationship is strong and positive
### -1 strong and ngetive relationship
### 0.5 +ve and moderate
### 0.1 -> weak and +ve
